```python
from datetime import datetime, timezone, timedelta

import pandas as pd

from research.zack import fetch_data_for_date, parse_json_blob

# Set today and tomorrow dates to 00:00
today_date = datetime.now().replace(hour=5, minute=0, second=0, microsecond=0) + timedelta(days=1)
tomorrow_date = today_date + timedelta(days=1)

# Fetch and parse data for today
content_str_today = fetch_data_for_date(today_date)
earnings_df_today = parse_json_blob(content_str_today) if content_str_today else None
earnings_df_today['Date'] = today_date.strftime('%Y-%m-%d')

# Fetch and parse data for tomorrow
content_str_tomorrow = fetch_data_for_date(tomorrow_date)
earnings_df_tomorrow = parse_json_blob(content_str_tomorrow) if content_str_tomorrow else None
earnings_df_tomorrow['Date'] = tomorrow_date.strftime('%Y-%m-%d')

# Combine dataframes for today and tomorrow
if earnings_df_today is not None and earnings_df_tomorrow is not None:
    earnings_df = pd.concat([earnings_df_today, earnings_df_tomorrow])
    print("Data for today and tomorrow extracted and DataFrame formed successfully.")
else:
    print("Failed to extract data.")

# Define today's and tomorrow's dates as strings for comparison
today_str = today_date.strftime('%Y-%m-%d')
tomorrow_str = tomorrow_date.strftime('%Y-%m-%d')
```

```python
# Apply the filtering conditions
today_action_df = earnings_df[
    (earnings_df['EPS Percentage'] != 0.0) &
    ((earnings_df['Date'] == today_str) & (earnings_df['AMC/BMO'].str.lower() == 'amc') |
     (earnings_df['Date'] == tomorrow_str) & (earnings_df['AMC/BMO'].str.lower() == 'bmo'))
]
today_action_df
```
