```python
import requests

redirect_uri = "https://dummy-url-to-redirect.gg/"

auth_params = {
    "response_type": "code",
    "client_id": client_id,
    "redirect_uri": redirect_uri,
    "scope": "openid profile"
}

# Make a GET request to the authorize endpoint
authorize_response = requests.get(authorize_url, params=auth_params)
```

```python
# print(json.dumps(response.json(), indent=4))
print(authorize_response.url)
```

```python
# Extract the authorization code from the response
auth_code = "dd76de43-ae3a-4e2d-b0e4-6a2c613fa0ce"  # 04:21 23Sep

# Make a POST request to the token endpoint to get the access token
token_params = {
    "grant_type": "authorization_code",
    "code": auth_code,
    "redirect_uri": redirect_uri
}

auth = (client_id, client_secret)
token_response = requests.post(token_url, data=token_params, auth=auth)

# Extract the access token from the response
access_token = token_response.json()["access_token"]

print("Access token:", access_token)
```
